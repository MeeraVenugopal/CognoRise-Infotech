**BREAST CANCER CLASSIFICATON**

Classify breast cancer tumors as malignant or benign using features extracted from mammograms.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/archive (1).zip')
df
```

```python
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
df
```

Change the values in output to its names

```python
df['diagnosis']=df['diagnosis'].map({'M':'Malignant','B':'Benign'})
df
```

```python
plt.pie(df['diagnosis'].value_counts(),labels=['Malignant','Benign'],colors=['hotpink','yellow'],autopct='%1.1f%%')
plt.legend()
```

Output Removal


```python
for i in df.columns:
  plt.figure(figsize=(6,6))
  sns.boxplot(df[i],color='green')
  plt.xlabel(i)
  plt.ylabel('Range')
```

```python
for i in df.columns[1:]:
  def iqr_method(df,i):
      q1=df[i].quantile(0.25)
      q2=df[i].quantile(0.75)
      IQR=q2-q1
      max_limit=q2+(1.5*IQR)
      min_limit=q1-(1.5*IQR)
      return max_limit,min_limit
  upper,lower=iqr_method(df,i)
  df[i]=np.where(df[i]>upper,upper,df[i])
  df[i]=np.where(df[i]<lower,lower,df[i])
```

```python
for i in df.columns[1:]:
  plt.figure(figsize=(6,6))
  sns.boxplot(df[i],color='green')
  plt.xlabel(i)
  plt.ylabel('Range')
```

Feature selection

```python
df1=df.drop('diagnosis',axis=1)
corr_matrix =df1.corr()
corr_matrix
```

Separating inputs and output columns


```python
x=df.drop('diagnosis',axis=1).values
x
```

```python
y=df['diagnosis'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

Normalisation

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

Model creation

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
knn=KNeighborsClassifier()
classifier=RandomForestClassifier(n_estimators=17,criterion='log_loss',random_state=0,max_features='sqrt')
model=DecisionTreeClassifier()
lst=[model,classifier,knn]
```

```python
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('Model is',i)
  print(y_pred)
  print('-'*100)
  print('The accuracy score of',i,'is',accuracy_score(y_test,y_pred))
  print('-'*100)
  print(confusion_matrix(y_test,y_pred),'is the confusion matrix of',i)
  print('-'*100)
  print(classification_report(y_test,y_pred),'is the classification report of',i)
  print('*'*100)
```

Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
param={'n_neighbors':[3,5,7,9,11,13],'weights':['uniform','distance']}
grid=GridSearchCV(knn,param,cv=10,scoring='accuracy')
grid.fit(x_train,y_train)
print(grid.best_params_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn1=KNeighborsClassifier(n_neighbors=9,weights='uniform')
knn1.fit(x_train,y_train)
y_pred2=knn1.predict(x_test)
print(accuracy_score(y_test,y_pred2))
```
