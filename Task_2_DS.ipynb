```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/CarPrice_Assignment.csv')
df
```

```python
df.drop(['car_ID'],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
categorical_columns = df.select_dtypes(include=['object']).columns

# Apply label encoding to all categorical columns
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
df
```

```python
for i in['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']:
  plt.figure(figsize=(6,6))
  sns.boxplot(df[i],color='red')
  plt.xlabel(i)
  plt.ylabel('Range')
```

```python
q1=df.quantile(0.25)
q2=df.quantile(0.75)
IQR=q2-q1
max_limit=q2+(1.5*IQR)
min_limit=q1-(1.5*IQR)
```

```python
df=pd.DataFrame(np.where(df>max_limit,max_limit,(np.where(df<min_limit,min_limit,df))),columns=df.columns)
```

```python
for i in['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']:
  plt.figure(figsize=(6,6))
  sns.boxplot(df[i],color='green')
  plt.xlabel(i)
  plt.ylabel('Range')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot=True)
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
r2=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
print("R2 Score:",r2)
print("Mean Squared Error:",mse)
```
